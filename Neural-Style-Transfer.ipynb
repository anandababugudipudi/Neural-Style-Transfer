###**Import the necessary packages**

```python
import tensorflow_hub as hub
import tensorflow as tf
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

###**Get the model**

```python
model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

###**Image Conversion**

```python
def load_image(img_path):
    img = tf.io.read_file(img_path)
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = img[tf.newaxis, :]
    return img
```

###**Load the Images**

```python
content_image = load_image('/content/baby_yoda.jpeg')
style_image = load_image('/content/monet.jpeg')
```

###**Check the Images**

```python
plt.imshow(np.squeeze(content_image))
```

```python
plt.imshow(np.squeeze(style_image))
```

```python
stylized_image = model(tf.constant(content_image), tf.constant(style_image))[0]
```

```python
plt.imshow(np.squeeze(stylized_image))
```

###**Write a function for returning the Stylized image**

```python
def stylize_images(content_image_path, style_image_path):
  content_image = load_image(content_image_path)
  style_image = load_image(style_image_path)
  stylized_image = model(tf.constant(content_image), tf.constant(style_image))[0]
  return stylized_image
```

```python
stylized_image = stylize_images('/content/Panda.jpeg', '/content/vangogh.jpeg')
plt.imshow(np.squeeze(stylized_image))
```
